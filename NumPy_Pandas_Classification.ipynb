**Task-5**

```python
from sklearn.datasets import load_iris
```

```python

iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['target'] = iris.target

df['binary_target'] = (df['target'] == 1).astype(int)

df.head()
```

```python
from sklearn.model_selection import train_test_split

X = df[iris.feature_names]
y = df['binary_target']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

(X_train.shape, X_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression


```

```python
model = LogisticRegression(max_iter=200)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
y_pred[:5]
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score

conf_matrix = confusion_matrix(y_test, y_pred)

accuracy = accuracy_score(y_test, y_pred)

precision = precision_score(y_test, y_pred)

recall = recall_score(y_test, y_pred)

f1 = f1_score(y_test, y_pred)

conf_matrix, accuracy, precision, recall, f1
```

**Confusion Matrix:**
The confusion matrix is a 2x2 matrix:
[[29, 3],
[ 9, 4]]






*   **True Negatives (TN)** = 29: Correctly predicted "not Versicolor".
*   **False Positives (FP)** = 3: Incorrectly predicted "Versicolor".
*   **False Negatives (FN)** = 9: Incorrectly predicted "not Versicolor".
*   **True Positives (TP)** = 4: Correctly predicted "Versicolor".








**Metrics:**
- **Accuracy**: 73.33%  Percentage of correct predictions out of all predictions.
- **Precision**: 57.14%  How many of the predicted Versicolor cases were actually Versicolor.
- **Recall**: 30.77%  How many of the actual Versicolor cases were correctly identified.
- **F1-Score**: 40%  The harmonic mean of precision and recall, balancing the trade-off between them.

These metrics show that the model has a decent accuracy, but recall is much lower, indicating that the model misses many actual Versicolor instances.
